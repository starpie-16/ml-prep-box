```python
# Import module and libraries

from datawise.engine import toolbox
import seaborn as sns
import pandas as pd


#---------------------------------------------------------------

df = sns.load_dataset('titanic')
tb = toolbox()
```

```python
# Data & Optimization

mem_before = df.memory_usage(deep=True).sum() / 1024**2
optimized_df = tb.optimize_memory(df)
mem_after = optimized_df.memory_usage(deep=True).sum() / 1024**2

print(f"Memory Usage Before: {mem_before:.2f} MB")
print(f"Memory Usage After: {mem_after:.2f} MB")
print(f"Efficiency: {(1 - mem_after/mem_before)*100:.1f}% reduction")
print("-" * 30)
```

```python
# Setup
tb.setup(
    data=optimized_df,
    target='survived',
    categorical_features=['pclass', 'sex', 'embarked']
)

# Display suggested strategies
print("\n--- SUGGESTED STRATEGIES ---")
tb.suggest()
```

```python
# Transformation

from IPython.display import display


#---------------------------------------------------------------


transformed_df = tb.apply_transform()

print("\n FINAL PROCESSED DATA (PREVIEW first 5 rows):")
display(transformed_df.head())

# Display the final general information
print("\n Final Info:")
transformed_df.info()
```
